###책 제목

```python
import requests
from bs4 import BeautifulSoup

#스크래핑할 주소
url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we'

#헤더 User-Agent 설정
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'}
data = requests.get(url,headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

###데이터 추출하기 (CSS 선택자 활용)

```python
title = soup.select("a.bo3")

#결과 list형태로 반환
print(title)
```

```python
#제목만 추출
for t in title:
    print(t.text.strip()) # 하나씩 꺼내서 text 출력
```

###책소개

```python
url = 'https://www.aladin.co.kr/shop/wproduct.aspx?ItemId=376765918'

data = requests.get(url,headers=headers)
soup = BeautifulSoup(data.text, 'html.parser')

print(soup)
```

```python
desc = soup.select("#db_introduction")

#결과 list형태로 반환
print(desc)
```

```python
import time

book_data_list = []

#책소개를 담는 함수 정의
def get_description(sub_soup):
    boxes = sub_soup.select(".Ere_prod_mconts_box")

    for box in boxes:
        ls = box.select_one(".Ere_prod_mconts_LS")
        if ls and any(k in ls.text for k in ["소개", "줄거리"]):
            content = box.select_one(".Ere_prod_mconts_R")
            if content:
                return content.get_text(" ", strip=True)
    return "설명을 찾을 수 없습니다."

#반복문으로 책소개 담기
for i, tag in enumerate(title[:10]):
    book_title = tag.text.strip()
    link = tag['href']

    try:
        res = requests.get(link, headers=headers)
        sub_soup = BeautifulSoup(res.text, 'html.parser')

        description = get_description(sub_soup)

        book_data_list.append({
            "rank": i + 1,
            "title": book_title,
            "description": description
        })

        print(f"[{i+1}/10] {book_title} 수집 성공")
        time.sleep(0.5) # 서버 부하 방지

    except Exception as e:
        print(f"[{i+1}/10] {book_title} 실패: {e}")
```

```python
print(f"상세 설명:\n{book_data_list[6]['description']}")
```

### 청킹

```python
def chunk_titles(data, chunk_size=10):
    # 리스트를 chunk_size 단위로 잘라서 새로운 리스트에 담음
    return [data[i:i + chunk_size] for i in range(0, len(data), chunk_size)]
```
